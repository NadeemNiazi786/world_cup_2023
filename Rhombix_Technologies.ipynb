```python
Q1 = combined["score"].quantile(0.25)
Q3 = combined["score"].quantile(0.75)

IQR = Q3 - Q1
\
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

score_outliers = combined[(combined["score"] < lower) | (combined["score"] > upper)]

print(score_outliers)
```

```python
# Verification of outliers for score column
# if max score < upper
# and min score > lower
# no outliers
# if no dots outside the box in boxplot then no outliers
print("Lower:", lower)
print("Upper:", upper)
print("Min score:", combined["score"].min())
print("Max score:", combined["score"].max())
```

```python
def find_outliers(column):
    Q1 = combined[column].quantile(0.25)
    Q3 = combined[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return combined[(combined[column] < lower) | (combined[column] > upper)]
```
